# Part - 1 Archived Tweet Analysis
In this notebook we import the dataframes created in the previous notebook and explore how valuable the sentiment analysers are in real world use.  There are several relationships to explore so I will go through them in no particular order.

```python
import pandas as pd
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pickle
import datetime
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pickle_in=open("df.pickle","rb")
df=pickle.load(pickle_in)
```

## Nominal Tweet Volume:
This will explore how the number of tweets about a particular stock since the previous close of tradeing correlates to the stock's performance the following trading day.

```python
test_range=np.linspace(20,2000,100)
results=pd.DataFrame([])
for j in test_range:
    # This isolates the trades with tweet volume greater than the threshold into a data frame called "big"
    big=df[df.num_stock_tweets>j]
    big.reset_index(drop=True,inplace=True)
    min_sentiment=.0000001
    x_v_date=big.trade_date
    x_v=big.stock_weighted_vader
    x_b=big.stock_weighted_blob
    y=big.stock_move
    fails_v,fails_b=[],[]
    vader_trades=pd.DataFrame([])
    blob_trades=pd.DataFrame([])
    # This creates a dataframe of the Vader trades
    for i in range(len(x_v)):
        if (x_v[i]<-min_sentiment) & (y[i]>min_sentiment):
            vader_trades=vader_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':-y[i]},index=[0]),ignore_index=True)
        elif (x_v[i]>min_sentiment) & (y[i]>min_sentiment):
            vader_trades=vader_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':y[i]},index=[0]),ignore_index=True)
        elif (x_v[i]>min_sentiment) & (y[i]<-min_sentiment):
            vader_trades=vader_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':y[i]},index=[0]),ignore_index=True)
        elif (x_v[i]<-min_sentiment) & (y[i]<-min_sentiment):
            vader_trades=vader_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':-y[i]},index=[0]),ignore_index=True)
        else:
            fails_v.append(y[i]) 
    # This creates a dataframe of the Blob trades
    for i in range(len(x_b)):
        if (x_b[i]<-min_sentiment) & (y[i]>min_sentiment):
            blob_trades=blob_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':-y[i]},index=[0]),ignore_index=True)
        elif (x_b[i]>min_sentiment) & (y[i]>min_sentiment):
            blob_trades=blob_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':y[i]},index=[0]),ignore_index=True)
        elif (x_b[i]>min_sentiment) & (y[i]<-min_sentiment):
            blob_trades=blob_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':y[i]},index=[0]),ignore_index=True)
        elif (x_b[i]<-min_sentiment) & (y[i]<-min_sentiment):
            blob_trades=blob_trades.append(pd.DataFrame({'date':x_v_date[i],'stock_move':-y[i]},index=[0]),ignore_index=True)
        else:
            fails_b.append(y[i]) 
    # This creates a dataframe for the real-world-return for Vader trades, hence "rwrv"
    rwrv=pd.DataFrame([])
    rwrv_days=big.trade_date.unique()
    for i in rwrv_days:
        isov=big[big.trade_date==i]
        vad=vader_trades[vader_trades.date==i]
        al_trades=df[df.trade_date==i]
        rwrv=rwrv.append(pd.DataFrame({'day':i,'vader':vad.stock_move.mean(),'iso_trades':isov.stock_move.mean(),
                                     'all_trades':al_trades.stock_move.mean(),
                                    'num_vader':len(vad),'num_all':len(al_trades)},index=[0]),ignore_index=True)  
    # This creates a dataframe for the real-world-return for Blob trades, hence "rwrb"
    rwrb=pd.DataFrame([])
    rwrb_days=big.trade_date.unique()
    for i in rwrb_days:
        isob=big[big.trade_date==i]
        blo=blob_trades[blob_trades.date==i]
        al_trades=df[df.trade_date==i]
        rwrb=rwrb.append(pd.DataFrame({'day':i,'blob':blo.stock_move.mean(),'iso_trades':isob.stock_move.mean(),
                                     'all_trades':al_trades.stock_move.mean(),
                                    'num_vader':len(blo),'num_all':len(al_trades)},index=[0]),ignore_index=True) 
    # This brings it all together
    results=results.append(pd.DataFrame({'min_tweets':j,'num_vader_trades':len(vader_trades),
                                         'vader_pct_correct':vader_trades.stock_move.gt(0).sum()/len(vader_trades),
                                         'blob_pct_correct':blob_trades.stock_move.gt(0).sum()/len(blob_trades),
                                        'iso_pct_up':big.stock_move.gt(0).sum()/len(big),
                                         'all_pct_up':df.stock_move.gt(0).sum()/len(df),
                                        'total_vader_return':rwrv.vader.sum(),
                                         'total_blob_return':rwrb.blob.sum(),
                                         'total_iso_return':rwrv.iso_trades.sum(),
                                        'total_bh':rwrv.all_trades.sum(),
                                        'avg_vader_ret':vader_trades.stock_move.mean(),
                                         'avg_blob_ret':blob_trades.stock_move.mean(),
                                        'avg_iso_ret':big.stock_move.mean(),
                                        'avg_bh_ret':df.stock_move.mean()},index=[0]),ignore_index=True)
bench_pct=results.all_pct_up[0]*100 # For easy use in later charts
bench_ret=results.total_bh[0]*100 # For easy use in later charts
bench_avg_ret=results.avg_bh_ret[0]*100 # For easy use in later charts
plt.figure(figsize=(10,5))
plt.plot(results.min_tweets,results.vader_pct_correct*100,label='Vader Predicted Trades')
plt.plot(results.min_tweets,results.blob_pct_correct*100,label='Blob Predicted Trades')
plt.plot(results.min_tweets,results.iso_pct_up*100,label=' Buy Isolated Trades')
plt.plot(results.min_tweets,results.all_pct_up*100,label='Buy All Trades')
plt.legend()
plt.xlabel('Minimum Number of Tweets')
plt.ylabel('Percent Correct')
plt.title('Percent of Trades Directionally Correct')
plt.figure(figsize=(10,5))
plt.plot(results.min_tweets,results.total_vader_return*100,label='Vader Predicted Trades')
plt.plot(results.min_tweets,results.total_blob_return*100,label='Blob Predicted Trades')
plt.plot(results.min_tweets,results.total_iso_return*100,label=' Buy Isolated Trades')
plt.plot([results.min_tweets[0],results.min_tweets[len(results)-1]],[results.total_bh[0]*100,results.total_bh[0]*100],label='Buy All Trades')
plt.legend()
plt.xlabel('Minimum Number of Tweets')
plt.ylabel("Total Return (Percentage)")
plt.title('Total Return')
# Note: the Buy All Trades line moves because it throws out days in which there was not sufficient tweeeting
plt.figure(figsize=(10,5))
plt.plot(results.min_tweets,results.avg_vader_ret*100,label='Vader Predicted Trades')
plt.plot(results.min_tweets,results.avg_blob_ret*100,label='Blob Predicted Trades')
plt.plot(results.min_tweets,results.avg_iso_ret*100,label=' Buy Isolated Trades')
plt.plot(results.min_tweets,results.avg_bh_ret*100,label='Buy All Trades')
plt.legend()
plt.xlabel('Minimum Number of Tweets')
plt.ylabel("Average Return Per Trade (Percentage)")
plt.title('Average Return Per Trade')
```

#### Top Chart
Both the Vader and Blob techniques are useful in predicting the direction of trades with greater accuracy than simply buying every trade and also are more accurate than buying all of the trades with tweets above the threshold.  This suggests that there is value in determining sentiment about a stock beyond simply isolating trades within a certain tweet range.  This holds true up until about the 1,100 tweet threshold which is where the sample size diminishes rapidly and likely leads to the breakdown of usefulness.

#### Middle Chart
We see that the only advantage found from actual trading is when the Vader algorithm is used on trades with a relatively low minimum tweet threshold.  Aside from that we would be better off simply buying the whole market.

#### Bottom Chart
Like in the middle chart we see that having a minimum tweet threshold below about 200 gives an advantage in average return per trade, however this time both the Vader and Blob algorithm show usefulness.  Minimum tweet thresholds above about 250 seem to be counter productive.

The Blob algorithm outperforms the benchmark in most cases but not by very much.  Importantly, for most of the spectrum the Vader trades outperform simply buying the trades with respectively high tweet volume, meaning that the sentiment analyser is adding value.

#### Middle Chart
Again we see that the Vader algorithm outperforms across the spectrum and adds increasing value as the relative tweet volume increases (distance between the blue and green line). Using Vader Sentiment on high relative tweet volume trades is clearly a winning strategy but its usefulness appears to fade beyond about 1.25 standard deviations.  This is because there are fewer trades which meet that threshold and thus the total return will diminish.

#### Bottom Chart
Here we see that the average return per trade continues to increase as the relative tweet volume increases.  This is as expected because the more tweets, the more insight we have into the market's sentiment about a particular stock.

```python
big=df[df.num_stock_tweets>(df.stock_num_tweet_mean+df.stock_num_tweet_std*1)]
big.reset_index(drop=True,inplace=True)
min_vader=.0000001
x_date=big.trade_date
x=big.stock_weighted_vader
y=big.stock_move
#x=big.stock_weighted_vader-big.sector_weighted_vader
#y=big.stock_move-big.sector_move
q1,q2,q3,q4,fails=[],[],[],[],[]
vader_trades=[]
vader_trades_df=pd.DataFrame([])
for i in range(len(x)):
    if (x[i]<-min_vader) & (y[i]>min_vader):
        q1.append(y[i])
        vader_trades.append(-y[i])
        vader_trades_df=vader_trades_df.append(pd.DataFrame({'date':x_date[i],'stock_move':-y[i]},index=[0]),ignore_index=True)
    elif (x[i]>min_vader) & (y[i]>min_vader):
        q2.append(y[i])
        vader_trades.append(y[i])
        vader_trades_df=vader_trades_df.append(pd.DataFrame({'date':x_date[i],'stock_move':y[i]},index=[0]),ignore_index=True)
    elif (x[i]>min_vader) & (y[i]<-min_vader):
        q3.append(y[i])
        vader_trades.append(y[i])
        vader_trades_df=vader_trades_df.append(pd.DataFrame({'date':x_date[i],'stock_move':y[i]},index=[0]),ignore_index=True)
    elif (x[i]<-min_vader) & (y[i]<-min_vader):
        q4.append(y[i])
        vader_trades.append(-y[i])
        vader_trades_df=vader_trades_df.append(pd.DataFrame({'date':x_date[i],'stock_move':-y[i]},index=[0]),ignore_index=True)
    else:
        fails.append(y[i])
total_accepted=len(q1)+len(q2)+len(q3)+len(q4)
print('Upper Left Quadrant: ',len(q1),' trades for ',round(-np.sum(q1)*100,2),'% return')
print('Upper Right Quadrant: ',len(q2),' trades for ',round(np.sum(q2)*100,2),'% return')
print('Lower Right Quadrant: ',len(q3),' trades for ',round(np.sum(q3)*100,2),'% return')
print('Lower Left Quadrant: ',len(q4),' trades for ',round(-np.sum(q4)*100,2),'% return')
print('Percent of Trades Predicted Directionally Correct: ',round((len(q2)+len(q4))/(len(q1)+len(q2)+len(q3)+len(q4))*100,2),'%')
print('Percent of Trades that went Up: ',round((len(q1)+len(q2))/(len(q1)+len(q2)+len(q3)+len(q4))*100,2),'%')
print('Average Weighted Vader Return: ',round(((np.sum(q2)-np.sum(q4))-(np.sum(q1)-np.sum(q3)))*100/total_accepted,3),'%')
print('Average High Tweet Volume Buy and Hold Return: ',round(((np.sum(q2)+np.sum(q4))+(np.sum(q1)+np.sum(q3)))*100/total_accepted,3),'%')
print('Average Return if all Trades are Bought: ',round(df.stock_move.mean()*100,2),'%')
plt.figure(figsize=(8,6))
plt.scatter(x,y*100,alpha=.4)
#plt.xlim(-.3,.3)
#plt.ylim(-.03,.03)
plt.axvline(0,color='red')
plt.axhline(0,color='red')
plt.xlabel('Weighted Vader Score')
plt.ylabel('Stock Move (Percent)')
plt.title('Weighted Vader Score vs Resulting Stock Move')

# rwr is a dataframe of the real-world-returns that creates each day's trades
rwr=pd.DataFrame([])
rwr_days=big.trade_date.unique()
for i in rwr_days:
    iso=big[big.trade_date==i]
    vad=vader_trades_df[vader_trades_df.date==i]
    al_trades=df[df.trade_date==i]
    rwr=rwr.append(pd.DataFrame({'day':i,'vader':vad.stock_move.mean(),'iso_trades':iso.stock_move.mean(),
                                 'all_trades':al_trades.stock_move.mean(),
                                'num_vader':len(vad),'num_all':len(al_trades)},index=[0]),ignore_index=True)
rwr=rwr.sort_values(by='day')
rwr.reset_index(drop=True,inplace=True)
rwr['sum_vader']=rwr.vader.cumsum()
rwr['sum_iso']=rwr.iso_trades.cumsum()
rwr['sum_all']=rwr.all_trades.cumsum()
rwr['diff']=rwr.sum_vader-rwr.sum_all

welch_results=stats.ttest_ind(vader_trades,y)
print('Welch Coefficient: ',round(welch_results[0],3))
print('welch p-value: ',round(welch_results[1],3))

plt.figure(figsize=(12,6))
plt.plot(rwr.day,rwr.sum_vader*100,label='Weighted Vader Predicted Trades')
plt.plot(rwr.day,rwr.sum_iso*100,label='High Tweet Volume Isolated Trades')
plt.plot(rwr.day,rwr.sum_all*100,label='All Trades')
plt.legend()
plt.ylabel('Cumulative Return (Percentage)')
plt.title('Performance of Portfolios')
plt.xticks(rotation=45)
ax=plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(14))
```

The upper chart is a plot of the Weighted Vader Score against the resulting stock move. The upper right and lower left quadrants are trades in which the Weighted Vader Score correctly predicted the direction of the stock move and the upper left and lower right quadrants are trades that were predicted incorrectly.  We can see from the sum of returns in each quadrant that correctly predicted trades have a greater magnitude of movement and thus there is profit potential.  The average per trade return for Weighted Vader predicted trades is almost three times that of high tweet volume trades and more than triple the average of all trades.  The percentage of trades that were predicted correcly was 55% vs the 50.7% benchmark of simply predicting that every trade would go up.  This suggests that there is value to be found in mining Twitter for market sentiments about particular stocks.
